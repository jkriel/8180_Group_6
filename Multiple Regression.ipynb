```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('Concord.csv', sep=',')
data.head()
```


### The data in this set is comprised of the following fields and theie respective description

>#### water81 : Cubic feet of household water use in 1981  
>#### water80 : Cubic feet of household water use in 1980
>#### water79 : Cubic feet of household water use in 1979
>#### income  : 1981 household income in $1000s
>#### educat  : Education of head of household, in years
>#### peop81  : Number of people living in household summer 1981
>#### cpeop   : Difference between peop81 and peop80 
>#### peop80  : Number of people living in household summer 1980
>#### retired : 1 if head of household is retired, 0 otherwise

```python
x = data.iloc[:,2:9].drop('water79', axis=1)

y = data['water81']

xPlusC = sm.add_constant(x)
results = sm.OLS(y,xPlusC).fit()
results.summary()
```

```python
x2 = data.iloc[:,2:9].drop(['water79','cpeop'], axis=1)

y = data['water81']

x2PlusC = sm.add_constant(x2)
results2 = sm.OLS(y,x2PlusC).fit()
print("With intercept:"+"\n",results2.summary(),"\n"*2)
results3 = sm.OLS(y,x2).fit()
print("Without intercept:"+"\n",results3.summary())
```
